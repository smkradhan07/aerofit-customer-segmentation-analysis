```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset (Assuming the file is 'aerofit_treadmill.csv')
aerofit = pd.read_csv('aerofit_treadmill.csv')
```

```python
# Marginal Probability of Products [cite: 7]
marginal_probabilities = aerofit['Product'].value_counts(normalize=True) * 100
print("Marginal Probabilities (%):\n", marginal_probabilities)
```

```python
# Conditional Probability (Gender and Product) [cite: 7]
gender_conditional = pd.crosstab(aerofit['Gender'], aerofit['Product'], normalize='index')
print("\nGender vs Product Probability:\n", gender_conditional)
```

```python
# Statistical Profiling by Product [cite: 8]
print("\nAverage characteristics per Product segment:\n")
print(aerofit.groupby('Product')[['Income', 'Miles', 'Fitness']].mean())
```
