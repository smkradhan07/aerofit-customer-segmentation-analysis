```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset (Assuming the file is 'aerofit_treadmill.csv')
aerofit = pd.read_csv('aerofit_treadmill.csv')
```

```python
# Missing Value Detection [cite: 6]
print("Missing values in dataset:\n", aerofit.isna().sum())
```

```python
# Outlier Detection for Income [cite: 6]
plt.figure(figsize=(12, 6))
sns.boxplot(data=aerofit, x='Income')
plt.title('Detection of Income Outliers')
plt.show()
```
