```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset (Assuming the file is 'aerofit_treadmill.csv')
aerofit = pd.read_csv('aerofit_treadmill.csv')
```

```python
# Univariate Analysis [cite: 83, 84, 90, 91, 97, 98]
plt.figure(figsize=(10, 5))
sns.histplot(aerofit['Age'], kde=True, color='blue')
plt.title('Distribution of Customer Age')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(aerofit['Income'], kde=True, color='green')
plt.title('Distribution of Annual Income')
plt.show()
```

```python
# Bivariate Analysis [cite: 106, 107, 113, 114, 120, 121]
plt.figure(figsize=(10, 6))
sns.boxplot(data=aerofit, x='Product', y='Income')
plt.title('Income Distribution by Product')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=aerofit, x='Product', y='Miles')
plt.title('Expected Weekly Miles by Product')
plt.show()
```
